```python
import numpy as np
import matplotlib.pyplot as plt
# Generate a sinusoidal signal
x = np.sin(np.linspace(0, 2*np.pi, 1000))
quant_freq = np.arange(5, 1000, 1)
q_err = []
```

```python
for frequency in quant_freq:
    quant_signal = np.round(x*frequency)/frequency
    q_err.append(np.mean((x-quant_signal)**2))

plt.plot(quant_freq, q_err,'y')
plt.xlabel('Quantization Frequency')
plt.ylabel('Quantization Error')
plt.show()
```
